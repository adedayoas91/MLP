```python
#import packages
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
%matplotlib inline
```

```python
#visualise data

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for c, m, in [('r', 'o'), ('b', '^'), ('g', '*')]:
    ax.scatter(features[:,0], features[:,1], features[:,2], label='3D scatterplot of features', c=c, marker = m)
ax.set_xlabel('feature 1')   
ax.set_ylabel('feature 2')
ax.set_zlabel('feature 3')
fig.savefig('3D scatter plot.png')
```
